```python
import train
from agents import PerfectAgent
from ttt_env import TicTacToeEnv

agent = train.train(50000)
perfect = PerfectAgent()
env = TicTacToeEnv()

wins = draws = losses = 0

for _ in range(200):
    env.reset()
    done = False

    while not done:
        action = agent.act(env.board, env.available_actions(), training=False)
        _, reward, done = env.step(action)

        if done:
            if reward == 1:
                wins += 1
            else:
                draws += 1
            break

        opp_action = perfect.act(env.board, env.available_actions())
        _, reward, done = env.step(opp_action)

        if done:
            losses += 1

print("Vs Perfect Strategy")
print("Wins:", wins)
print("Draws:", draws)
print("Losses:", losses)
```
